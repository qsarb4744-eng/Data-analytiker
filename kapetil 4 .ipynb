```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn import tree

# 1. Ladda datasetet
iris = load_iris()
X = iris.data
y = iris.target

# 2. Skapa modellen
modell = DecisionTreeClassifier(random_state=1)

# 3. Träna modellen
modell.fit(X, y)

# 4. Visualisera beslutsträdet
plt.figure(figsize=(14,9))

tree.plot_tree(
    modell,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    rounded=True,
    impurity=False,   
    proportion=True   
)

plt.title("Beslutsträd med proportioner", fontsize=15)
plt.show()
```

```python
##13
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Ladda data
df = pd.read_excel("hr_employee_data.xlsx")

# 2. Dela upp X och y
X = df.drop("left", axis=1)
y = df["left"]

# 3. Omvandla kategoriska variabler
X = pd.get_dummies(X)

# 4. Dela upp i träning och test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 5. Skapa modell
model = DecisionTreeClassifier()

# 6. Träna modellen
model.fit(X_train, y_train)

# 7. Gör prediction
y_pred = model.predict(X_test)

# 8. Utvärdera modellen
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
##14
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Ladda data
iris = load_iris()
X, y = load_iris(return_X_y=True, as_frame=True)

# 2. Välj två variabler
X = X[['sepal length (cm)', 'sepal width (cm)']]

# 3. Dela data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 4. Skapa modell
model = DecisionTreeClassifier()

# 5. Träna modellen
model.fit(X_train, y_train)

# 6. Prediction
y_pred = model.predict(X_test)

# 7. Utvärdering
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# 8. Enkel visualisering
plt.scatter(X_train['sepal length (cm)'], X_train['sepal width (cm)'], c=y_train)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Träningsdata")
plt.show()
```
